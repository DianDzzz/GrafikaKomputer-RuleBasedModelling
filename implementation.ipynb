```python
import math
```

```python
class ProceduralArtist:
    def __init__(self, width=800, height=600):
        self.width = width
        self.height = height
        self.elements = []  # Menyimpan semua bentuk SVG
        self.stack = []     # Untuk menyimpan 'State' (posisi/sudut)

    # --- ACTION FUNCTIONS (Perintah Langsung) ---
    
    def draw_line(self, x1, y1, x2, y2, color="black", thickness=2):
        line = f'<line x1="{x1}" y1="{y1}" x2="{x2}" y2="{y2}" stroke="{color}" stroke-width="{thickness}" />'
        self.elements.append(line)

    def draw_rect(self, x, y, w, h, color="rgba(100, 150, 250, 0.5)"):
        rect = f'<rect x="{x}" y="{y}" width="{w}" height="{h}" fill="{color}" stroke="blue" />'
        self.elements.append(rect)

    def draw_circle(self, cx, cy, r, color="red"):
        circle = f'<circle cx="{cx}" cy="{cy}" r="{r}" fill="{color}" />'
        self.elements.append(circle)

    # --- EXPORT TO HTML ---
    
    def save_to_html(self, filename="procedural_model.html"):
        svg_content = "\n".join(self.elements)
        html_template = f"""
        <html>
        <body>
            <h1>Procedural Model Output</h1>
            <svg width="{self.width}" height="{self.height}" style="border:1px solid #ccc; background: #fafafa">
                {svg_content}
            </svg>
        </body>
        </html>
        """
        with open(filename, "w") as f:
            f.write(html_template)
        print(f"Model berhasil dibuat! Buka file '{filename}' di browser kamu.")
```

```python
# =========================================================
# JANTUNG RULE-BASED MODELLING
# =========================================================

def generate_model():
    artist = ProceduralArtist()

    # 1. AXIOM (Kondisi Awal)
    start_x, start_y = 400, 550
    
    # 2. RULES (Aturan Pertumbuhan/Pembagian)
    
    # --- CONTOH RULE 1: L-System (Pertumbuhan Cabang) ---
    def rule_tumbuh(x, y, angle, length, depth):
        if depth <= 0:
            artist.draw_circle(x, y, 3, "green") # Buat daun di ujung
            return

        # Aksi langsung: Hitung koordinat tujuan
        nx = x + length * math.sin(math.radians(angle))
        ny = y - length * math.cos(math.radians(angle))
        
        # Tambahkan garis ke list elemen
        artist.draw_line(x, y, nx, ny, "brown", depth)
        
        # Rekursi: Panggil aturan lagi dengan sudut berbeda
        # MODIFIKASI DI SINI: Ubah sudut atau jumlah cabang
        rule_tumbuh(nx, ny, angle - 25, length * 0.7, depth - 1)
        rule_tumbuh(nx, ny, angle + 25, length * 0.7, depth - 1)

    # --- CONTOH RULE 2: Shape Grammar (Pembagian Ruang Gedung) ---
    def rule_gedung(x, y, w, h, floors):
        if floors <= 0:
            artist.draw_rect(x, y - 20, w, 20, "gray") # Atap sederhana
            return
        
        # Aksi langsung: Gambar satu lantai
        artist.draw_rect(x, y - 40, w, 40)
        
        # MODIFIKASI DI SINI: Tambah jendela atau shape baru di setiap lantai
        if w > 30:
            artist.draw_circle(x + w/2, y - 20, 5, "yellow") # Jendela tengah
            
        # Rekursi: Tumpuk lantai ke atas
        rule_gedung(x, y - 40, w, h, floors - 1)

    # --- EKSEKUSI PERINTAH ---
    
    # Jalankan Rule Tumbuh (Pohon) di sebelah kiri
    rule_tumbuh(200, 500, 0, 100, 6)
    
    # Jalankan Rule Gedung (Shape Grammar) di sebelah kanan
    rule_gedung(500, 500, 100, 0, 8)

    # 3. FINALISASI
    artist.save_to_html()
```

```python
generate_model()
```
